K means Method:

```python
def calculate_dist(x1, x2):
    """Calculate the Euclidean distance between two points."""
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def update_centroid(image_indexes, data_set):
    """Update centroid by averaging the points in a cluster."""
    if len(image_indexes) == 0:
        return np.zeros(data_set[0][0].shape, dtype=np.float64)  # Avoid division by zero

    centroid = np.zeros_like(data_set[0][0], dtype=np.float64)  # Adjust dimension dynamically
    for idx in image_indexes:
        centroid += data_set[idx][0]  # Sum all feature vectors
    centroid /= len(image_indexes)  # Get mean

    return centroid
```

```python
from scipy import stats
def K_means_cluster(X_train , y_train):
  # Number of clusters
  K = 250
  clusters = []
  # Convert all columns of X_train to numeric, replacing errors with NaN
  X_train = X_train.apply(pd.to_numeric, errors='coerce')

  # Drop rows with NaN values after conversion
  X_train = X_train.dropna()

  # Reset index after dropping rows
  X_train = X_train.reset_index(drop=True)

  # Define K random points as initial cluster centroids
  feature_dim = X_train.shape[1]  # Get the actual feature dimension
  for i in range(K):
      point = np.random.uniform(-15, 15, size=(feature_dim,)).astype(np.float64)
      clusters.append([point, []])

  # Prepare dataset as (features, label) tuples
  # Use .values to get the underlying NumPy array of the DataFrame
  data_set = [[feature.astype(np.float64), int(label)] for feature, label in zip(X_train.values, y_train.values)]

  # K-means clustering (200 iterations)
  for iteration in tqdm(range(200), desc="Clustering Iterations"):
      # Clear previous cluster assignments
      for cluster in clusters:
          cluster[1] = []

      # Assign each data point to the nearest centroid
      for i, (features, _) in enumerate(data_set):
          distances = [calculate_dist(cluster[0], features) for cluster in clusters]
          cluster_index = np.argmin(distances)  # Index of closest cluster
          clusters[cluster_index][1].append(i)  # Assign data point index

      # Update centroids based on current cluster assignments
      for i in range(K):
          if len(clusters[i][1]) > 0:  # Only update if cluster is not empty
              clusters[i][0] = update_centroid(clusters[i][1], data_set)

  # Assign labels to clusters
  cluster_labels = []
  for i in range(K):
      labels_in_cluster = [data_set[idx][1] for idx in clusters[i][1]]
      if labels_in_cluster:  # Ensure cluster is not empty
          mode_label = stats.mode(labels_in_cluster, keepdims=True).mode[0]  # Extract mode correctly
      else:
          mode_label = -1  # Default label if cluster is empty
      cluster_labels.append(mode_label)
  return clusters,cluster_labels
```

```python
from tqdm import tqdm
```

```python
# Testing phase
from sklearn.metrics import accuracy_score
clusters_pca , cluster_labels_pca=K_means_cluster(X_train_pca , y_train_pca)
clusters_cnn , cluster_labels_cnn=K_means_cluster(X_train_cnn, y_train_cnn)

predictions_pca = [] # Create a list to store predictions_pca
for index in tqdm(range(X_test_pca.shape[0]), total=len(y_test_pca), desc="Testing Phase"): # Iterate over rows
    features = X_test_pca.iloc[index].astype(np.float64)  # Ensure correct type by using iloc to get row data
    distances = [calculate_dist(cluster[0], features) for cluster in clusters_pca]
    cluster_index = np.argmin(distances)  # Find nearest cluster
    predicted_label = cluster_labels_pca[cluster_index]  # Assign label based on cluster
    predictions_pca.append(predicted_label)

# Compute accuracy
accuracy = accuracy_score(y_test_pca, predictions_pca) # Pass predictions_pca to accuracy_score
print(f'Validation Accuracy for PCA features : {accuracy * 100:.2f}%')

# Testing for Cnn features:
predictions_cnn = [] # Create a list to store predictions_cnn
for index in tqdm(range(X_test_cnn.shape[0]), total=len(y_test_cnn), desc="Testing Phase"): # Iterate over rows
    features = X_test_cnn.iloc[index].astype(np.float64)  # Ensure correct type by using iloc to get row data
    distances = [calculate_dist(cluster[0], features) for cluster in clusters_cnn]
    cluster_index = np.argmin(distances)  # Find nearest cluster
    predicted_label = cluster_labels_cnn[cluster_index]  # Assign label based on cluster
    predictions_cnn.append(predicted_label)

# Compute accuracy
accuracy = accuracy_score(y_test_cnn, predictions_cnn) # Pass predictions_pca to accuracy_score
print(f'Validation Accuracy for CNN features : {accuracy * 100:.2f}%')
```
